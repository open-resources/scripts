```python
import pathlib
from bs4 import BeautifulSoup
import shutil
```

```python
path = pathlib.Path('../../edx_bank/library102/problem/test1_p9_vF.xml') # something with images

soup = BeautifulSoup(path.read_text(),'xml')
```

```python
# set the database

image_db = []
ends = []

for f in pathlib.Path('/Users/firasm/Downloads/course').glob('**/*'):
    ends.append(f.suffix)
    
    if f.suffix in ['.PNG','.png','.jpg','.jpeg','.gif']:
        image_db.append(f)
        
print(len(image_db),'\n\n',set(ends))
```

```python
for f in pathlib.Path('../../edx_bank/').glob('**/*.xml'):
    
    soup = BeautifulSoup(f.read_text(),'xml')
    
    if 'library' in f.name:
        continue
    else:
        #print(f)
        try: 
            image = pathlib.Path(soup.find_all('img')[0].get('src')).name
            
            
            found_image = [src for src in image_db if image in str(src)]
            
            if found_image:
                [shutil.copy2(src,f.parent / image) for src in found_image]
                print(image,'Success')
            else:
                print(image,': IMAGE NOT FOUND!!!!')
                raise NotImplementedError

        except IndexError: # No images in the problem
            image = 'undefined'
```
